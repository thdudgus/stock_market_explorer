```python
import pandas as pd

def get_stock_info():
    # KIND에서 상장법인목록 다운로드
    base_url =  "http://kind.krx.co.kr/corpgeneral/corpList.do"    
    method = "download"
    url = "{0}?method={1}".format(base_url, method)   
    
    # HTML 표 데이터를 읽어오기 (euc-kr 인코딩 필수)
    df = pd.read_html(url, header=0, encoding='euc-kr')[0]
    
    # 종목코드를 6자리 문자열로 포맷팅 (예: 5930 -> '005930')
    df['종목코드']= df['종목코드'].apply(lambda x: f"{x:06}")     
    
    return df

# 데이터 가져오기
df = get_stock_info()

# 데이터 상위 5개 행 확인
display(df.head())
```
